```python
def convert_to_experiment_params(A,w,T,Omega,phi,gamma):
    A_exp = A
    Omega_exp = Omega
    w_exp = w/Omega
    phi_exp = w*T/Omega/2+phi
    gamma_exp = gamma/Omega
    return A_exp, w_exp, T, Omega_exp, phi_exp, gamma_exp
```

```python
A, w, phi, T, gamma = [0.7049733914655495, 6.553990593919026,-4.7216575,7.65, 0.0]
Omega = 1*np.pi*2
convert_to_experiment_params(A,w,T,Omega,phi,gamma)
```
